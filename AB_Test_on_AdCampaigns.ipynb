#### In a year how many days did we observe low and high number of conversations ?

#### To assess this we need to spli the conversions for both the campigns into categories to identify low and high conversions

```python
def create_conversion_categories(conversion_col):
  category=[]
  for conversion in campaign_df[conversion_col]:
    if conversion <6:
       category.append('less than 6')
    elif 6<= conversion <11:
       category.append('6-10')
    elif 11<= conversion <16:
       category.append('10-15')
    else:
      category.append('more than 15')
  return category


campaign_df['Facebook Conversion Category']= create_conversion_categories("Facebook Ad Conversions")
campaign_df['AdWords Conversion Category']= create_conversion_categories("AdWords Ad Conversions")
```

## Hypothesis Testing

### Hypothesis Statement: Advertising on Facebook results in higher number of conversions compared to the adverting on adwords platform.

### Null Hypothesis: Number of conversions from adwords is greater than or equal to number of conversions from facebook ads.
H
0
​

### Alternate Hypothesis: Number of conversions from facebook are greater than adwords conversions
H
1

​> μ
AdWords
​


```python
print("Facebook Mean Conversion: ", round(campaign_df['Facebook Ad Conversions'].mean(), 2))
print("AdWords Mean Conversion': ", round(campaign_df['AdWords Ad Conversions'].mean(), 2))
```

### The mean number of conversions of the facebook ad campaigns is significantly higher than the mean of the adwords conversions by  adwords campaign. Suggesting that Facebook ads are more effective in encoraging users to buy the products.

```python
import scipy .stats as st

t_stats, p_value= st.ttest_ind(a=campaign_df['Facebook Ad Conversions'], b=campaign_df['AdWords Ad Conversions'], equal_var= False) # ttest_ind- is used for independent two-sample t-tests, where the goal is to test if the means of two independent samples are significantly different.
# equal_var: A boolean flag indicating whether the two groups are assumed to have equal variances. False: Does not assume equal variances.
print("T-statistics:", t_stats)
print("P-value:", p_value)

if p_value <0.05:
  print("\np-value is less than significance value, Thus we reject null hypothesis")
else:
  print("\np-value is more than significance value, Thus we Accept null hypothesis")
```

### Here we can see that the T-statistics is a large number, meaning that there is a significant difference between means of  both the groups.

### p value is less than the significance level i.e 0.05 thus we reject the Null Hypothesis. Thus smaller p-value indicates that there is stronger evidence to reject the null hypothesis.

### Thus Facebook ads indeed has better conversions than adwords. And we suggest stakeholders to inverst more in Facebook ads to make more revenue.

# Regression Analysis

### Now that we know that Facebook ads yeild better results we can then use regression analysis to forecast the conversions in the future based on the clicks.

## How many facebook ad conversions can we expect given a certain number of clicks?

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

X= campaign_df[['Facebook Ad Clicks']] #Independent variable
y= campaign_df[['Facebook Ad Conversions']] # Dependent variable

reg= LinearRegression()
reg.fit(X,y)
pred= reg.predict(X)

r2= r2_score(y, pred)
mse= mean_squared_error(y, pred)

print("R2 score:", r2)
print("mean squared error", mse)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=campaign_df['Facebook Ad Clicks'], y= campaign_df['Facebook Ad Conversions'])
plt.plot(campaign_df['Facebook Ad Clicks'], pred, color="r", label="best fit line")
plt.title("Regression Analysis: Prediction of conversions to clicks")
plt.legend()
plt.show()
```

```python
print(f'for {50} clicks, expected Conversion: {round(reg.predict([[50]])[0][0],2)}')
print(f'for {80} clicks, expected Conversion: {round(reg.predict([[100]])[0][0],2)}')
```

### Facebook Ad Campign metrics over time

```python
campaign_df["Facebook Click-Through Rate (Clicks / View)"]= campaign_df["Facebook Click-Through Rate (Clicks / View)"].apply(lambda x: float(x[:-1]))
campaign_df["Facebook Conversion Rate (Conversions / Clicks)"]= campaign_df["Facebook Conversion Rate (Conversions / Clicks)"].apply(lambda x: float(x[:-1]))
campaign_df["Facebook Cost per Click (Ad Cost / Clicks)"]= campaign_df["Facebook Cost per Click (Ad Cost / Clicks)"].apply(lambda x: float(x[1:]))
campaign_df["Cost per Facebook Ad"]= campaign_df["Cost per Facebook Ad"].apply(lambda x: float(x[1:]))
```

#### CPC (Cost per conversion)

CPC is a measure of cost effectiveness and profitability of ad campaign in this case. We can understand how much they are spending to obtain each conversion.

```python
monthly_df= campaign_df_fb.groupby('month')[['Facebook Ad Conversions', 'Cost per Facebook Ad']].sum()
monthly_df['Cost per conversion']= monthly_df['Cost per Facebook Ad']/monthly_df['Facebook Ad Conversions']
plt.figure(figsize=(10,4))
plt.title("Cost per Conversion")
plt.plot(Month_names, monthly_df['Cost per conversion'], "-o")
plt.show()
```

#### May and November months show lowers CPC, meaning they spend less for more number of conversions of all. February had highest CPC meaning that they spent most for least number of conversions. We can recommend to allocate more advertising budget to months with lower CPC like May and Nov to maximize the Return on inverstment.
